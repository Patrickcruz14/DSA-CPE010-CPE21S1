```
list1 = [26 + 49 + 98 + 87 + 62 + 75]
total_sum = sum(list1)

if total_sum % 2 == 0:
    print("the sum of the number is even:", total_sum)
else:
    print("the sum of the number is odd:", total_sum)
```
